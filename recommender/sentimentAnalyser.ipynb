```python
# import the libraries

from textblob import TextBlob
from wordcloud import WordCloud
from textblob import classifiers
import pandas as pd
import csv
import math

import nltk
nltk.download('punkt')
```

```python
with open('../data/data.csv', 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    output = list(reader)
    
train_data_count = math.floor((len(output) * 0.8))
train_data = []
test_data = []

for datapoint in output[1:train_data_count]:
    train_data.append(tuple(datapoint))
    
for datapoint in output[train_data_count:-1]:
    test_data.append(tuple(datapoint))
```

```python
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

def getPolarity(text):
    return TextBlob(text).sentiment.polarity

# df['Subjectivity'] = df['text_column']
# df = test_csv_data
# df['Subjectivity'] = df['Sentence'].apply(getSubjectivity)
# df['Polarity'] = df['Sentence'].apply(getPolarity)
```
